```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/House Price India.csv')
df.shape
```

```python
df.head
```

```python
df.tail
```

**Univariate Analysis**

```python
plt.hist(df['living area'],bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram for living area')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.kdeplot(df['living area'], shade=True, color='orange')
plt.title('Density Plot of living area ')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
sns.boxplot(x=df['living area'], color='green')
plt.title('Box Plot of living area')
plt.xlabel('Value')
plt.show()
```

```python
category_counts = df['living area'].value_counts()
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of living area')
plt.axis('equal')
plt.show()
```

```python
sns.barplot(x='living area', y='lot area', data=df, palette='viridis')
plt.title('Bar Plot of living area vs lot area')
plt.xlabel('Value')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Bi - Variate Analysis

```python
correlation_matrix = np.corrcoef(df['living area'], df['lot area'])
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap between living area and lot area')
plt.show()
```

**Multivariate Analysis**

```python
sns.pairplot(df[['living area',  'lot area', 'lot_area_renov', 'number of floors']], diag_kind='Built Year')
plt.suptitle('Pairwise Scatter Plot of Multiple Variables')
plt.show()
```

```python
correlation_matrix = df[['living area', 'lot area', 'number of floors', 'lot_area_renov']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix Heatmap of Multiple Variables')
plt.show()
```

Perform descriptive statistics on the dataset

```python
import pandas as pd
summary_stats = df.describe()
median = df.median()
variance = df.var()
skewness = df.skew()
kurtosis = df.kurt()
summary_df = pd.DataFrame({
    'Count': summary_stats.loc['count'],
    'Mean': summary_stats.loc['mean'],
    'Median': median,
    'Std Dev': summary_stats.loc['std'],
    'Variance': variance,
    'Skewness': skewness,
    'Kurtosis': kurtosis
})
print(summary_df)
```
